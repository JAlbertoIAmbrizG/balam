```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_prod_mex=pd.read_csv('Producción México.csv', index_col=0)
df_prod_mex
```

```python
df_prod_mex.info()
```

```python
import datetime
```

```python
balam=df_prod_mex.loc[df_prod_mex["CAMPO_SIH"]=='BALAM']
```

```python
fig, ax = plt.subplots(1, figsize=(20, 12))
plt.style.use('seaborn-darkgrid')
plt.scatter(balam['FECHA'], balam['PETROLEO_MBD'])
```

```python
balam
```

```python
balam.drop('FECHA', axis=1, inplace=True)
balam.drop('CAMPO_OFICIAL', axis=1, inplace=True)
balam.drop('CAMPO_SIH', axis=1, inplace=True)
balam.drop('UBICACION', axis=1, inplace=True)
balam.drop('PROVINCIA_PETROLERA', axis=1, inplace=True)
```

```python
balam
```

# <font color='green'>Data normalization</font>

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler(feature_range=(0, 1))
training_set_scaled = sc.fit_transform(balam)
training_set_scaled
```

# <font color='green'>Prepare the data structure for the NN format</font>

```python
X=training_set_scaled[:,1:6]
Y=training_set_scaled[:,0]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[5]),
    layers.Dense(10, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse', 'accuracy', 'mean_absolute_error'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
history = model.fit(X, Y, epochs=100, validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [PETROLEO_MBD]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$PETROLEO_MBD^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.legend()
  plt.show()


plot_history(history)
```

```python
fig, ax = plt.subplots(1, figsize=(20, 12))
plt.plot(Y)
plt.plot(model.predict(X))
plt.legend(['Real values', 'Predicted values'], loc='upper right')
```

```python
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(X, Y, epochs=100,
                    validation_split = 0.2, verbose=0, callbacks=[early_stop])

plot_history(history)
```

```python
fig, ax = plt.subplots(1, figsize=(20, 12))
plt.plot(Y)
plt.plot(model.predict(X))
plt.legend(['Real values', 'Predicted values'], loc='upper right')
```
